```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dat_e = np.genfromtxt("build/current_electron_3MeV.csv",delimiter=',')
dat_r = np.genfromtxt("build/reset_electron_3MeV.csv",delimiter=',')
```

```python
dat_r
```

```python
ElectronCharge_ = 1.60217662e-19
```

```python
plt.figure(figsize=(10,7))

Y = dat_e[:,1]

plt.plot(dat_e[:,0],Y, color='k', lw=3, label='Current profile')

plt.step(dat_r[:,1], dat_r[:,0]*6250,  
         where='post', color='darkmagenta',lw=3, label="C++ RTD")

# plt.step(dat_r[1::2][0:9], ((dat_r[0::2][0:9]*6250*ElectronCharge_)/10e-9)*1e9,  
#          where='post', color='darkmagenta',lw=3, label="C++ RTD")

# plt.step(dat_d[1::2][0:8], ((dat_d[0::2][0:8]*6250*ElectronCharge_)/10e-9)*1e9,  
#          where='post', color='darkblue',lw=3, label="C++ RTD 50ns dead \n w charge loss")

# plt.plot([X_res[-1],30e3], [Y_res[-1],Y_res[-1]], color='darkmagenta',lw=3)

plt.legend(loc="upper left", fontsize=18)
plt.ylabel("Current [nA]", fontsize=18)
plt.yticks(fontsize=16)

plt.xlabel("Time [ns]", fontsize=18)
plt.xticks(fontsize=16)
plt.grid()

# plt.xlim(3000,8000)
# plt.xlim(4e3,7e3)

plt.tight_layout()
# plt.savefig("current_profile.pdf")
plt.show()
```

```python
dat_e = np.genfromtxt("current_electron_3MeV.csv",delimiter=',')
dat_r = np.genfromtxt("reset_electron_3MeV.csv",delimiter=',')
dat_d = np.genfromtxt("reset_dead_electron_3MeV.csv",delimiter=',')
```

```python
print(len(dat_d[1::2]))
print(len(dat_d[0::2]))
```

```python
plt.figure(figsize=(10,7))

plt.plot(dat_e[1::2][0:2000],dat_e[0::2][0:2000], color='k', lw=3, label='Current profile')


plt.step(dat_r[1::2][0:9], ((dat_r[0::2][0:9]*6250*ElectronCharge_)/10e-9)*1e9,  
         where='post', color='darkmagenta',lw=3, label="C++ RTD")

plt.step(dat_d[1::2][0:8], ((dat_d[0::2][0:8]*6250*ElectronCharge_)/10e-9)*1e9,  
         where='post', color='darkblue',lw=3, label="C++ RTD 50ns dead \n w charge loss")

# plt.plot([X_res[-1],30e3], [Y_res[-1],Y_res[-1]], color='darkmagenta',lw=3)

plt.legend(loc="upper left", fontsize=18)
plt.ylabel("Current [nA]", fontsize=18)
plt.yticks(fontsize=16)

plt.xlabel("Time [ns]", fontsize=18)
plt.xticks(fontsize=16)
plt.grid()

plt.xlim(3000,8000)
# plt.xlim(4e3,7e3)

plt.tight_layout()
plt.savefig("current_profile.pdf")
plt.show()
```
