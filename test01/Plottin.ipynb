```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
RandomUniform = np.loadtxt("RandomUniform.txt")
RandomNormal = np.loadtxt("RandomNormal.txt")
```

```python
plt.figure(figsize=(7,7))
plt.hist(RandomUniform,bins=50)

#plt.semilogy()
plt.show()
```

```python
## fits the gaussian
hist, bin_edges = np.histogram(RandomNormal,bins=100)
bin_centers = (bin_edges[1:] + bin_edges[:-1])/2

def gaus(x,a,x0,sigma):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))
mean = 10
sigma=2
popt,pcov = curve_fit(gaus,bin_centers,hist,p0=[1,mean,sigma])
print("Height = ",popt[0])
print("Mean   = ",popt[1])
print("Sigma  = ",popt[2])
```

```python
plt.figure(figsize=(7,7))
plt.hist(RandomNormal,bins=100)
plt.plot(bin_centers,gaus(bin_centers,*popt),label='fit',color='r',linewidth=2)

plt.show()
```

```python
test = np.loadtxt("test.txt")
```

```python
## fits the gaussian
bins = np.arange(50)
hist, bin_edges = np.histogram(test,bins=bins)
bin_centers = (bin_edges[1:] + bin_edges[:-1])/2

def expo(x,a,tau):
    return a*np.exp(-x/tau)
mean = 10
sigma=2
popt,pcov = curve_fit(expo,bin_centers,hist,p0=[500,10])
print("Height = ",popt[0])
print("tau   = ",popt[1])
```

```python
plt.figure(figsize=(7,7))
plt.hist(test,bins=bins)
plt.plot(bin_centers,expo(bin_centers,*popt),label='fit',color='r',linewidth=2)

plt.show()
```
